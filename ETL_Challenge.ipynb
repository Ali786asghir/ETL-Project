```python
# Import Libraries
import pandas as pd
import sqlalchemy as sql
```

```python
# EVP Location File to Load 
file_to_load = ".\Resources\Electric_Vehicle_Population_Data.csv"

# Read Electric_Vehicle_Population_Data File and store into Pandas data frame
WA_EVP_Location_data = pd.read_csv(file_to_load)

WA_EVP_Location_data.head()
```

```python
#Refine Dataframe
WA_EVP_Location_clean_data = WA_EVP_Location_data[["VIN (1-10)", "County", "City", "ZIP Code", "Model Year", "Make", "Model", "Electric Range"]]
WA_EVP_Location_data

WA_EVP_Location_clean_data.head()
```

```python
# Save the WA_EVP_Location_clean_data dataframe as a csv
WA_EVP_Location_clean_data.to_csv('.\\Resources\WA_EVP_Location.csv', index = "False")
```
